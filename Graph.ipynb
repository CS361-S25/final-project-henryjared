```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def create_prop_temp_lum_graph(df, plots, title):

    df = df[(df['t'] >= 50000) & (df['t'] <= 6050000)]

    fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

    # Top plot: Proportion of Daisies
    if plots[0]:
        axs[0].plot(df['L'], df['a_b'], label='Black Daisies', color='black')

    if plots[1]:
        axs[0].plot(df['L'], df['a_w'], label='White Daisies', color='black', linestyle=':')

    if plots[2]:
        axs[0].plot(df['L'], df['a_g'], label='Gray Daisies', color='black', linestyle='dashdot')
        
    axs[0].set_xlim([0, 2])
    axs[0].set_ylim([0, 0.8])
    axs[0].set_yticks([.2, .4, .6])
    axs[0].legend(loc='upper center', ncols=3)
    axs[0].set_ylabel("Area %", size=13)


    axs[1].plot(df['L'], df['temp'], color='black')
    axs[1].set_xlim([0, 2])
    axs[1].set_ylim([0, 70])
    axs[1].set_yticks([20, 40, 60])
    axs[1].set_xlabel("Solar Luminosity", size=13)
    axs[1].set_ylabel("Temperature (Â°C)", size=13)


    plt.tight_layout()
    plt.xlim([0.6, 1.7])
    plt.savefig(title, dpi=600)
    plt.show()
```

```python
black_df = pd.read_csv("./data/black.csv")
black_df.name = "./figures/Black_Flat.png"

white_df = pd.read_csv("./data/white.csv")
white_df.name = "./figures/White_Flat.png"

gray_df = pd.read_csv("./data/gray.csv")
gray_df.name = "./figures/Gray_Flat.png"

white_black_df = pd.read_csv("./data/black_and_white.csv")
white_black_df.name = "./figures/White_Black_Flat.png"

white_black_gray_df = pd.read_csv("./data/white_black_and_gray.csv")
white_black_gray_df.name = "./figures/White_Black_Gray_Flat.png"

include_flat_plots = {
    black_df.name: [True, False, False],
    white_df.name: [False, True, False],
    gray_df.name: [False, False, True],
    white_black_df.name: [True, True, False],
    white_black_gray_df.name: [True, True, True]
}

flat_world_dfs = [black_df, white_df, gray_df, white_black_df, white_black_gray_df]

for df in flat_world_dfs:
    create_prop_temp_lum_graph(df, include_flat_plots[df.name], df.name)
```

```python
black_round_df = pd.read_csv("./data/black_round.csv")
black_round_df.name = "./figures/Black_Round.png"

white_round_df = pd.read_csv("./data/white_round.csv")
white_round_df.name = "./figures/White_Round.png"

white_black_round_df = pd.read_csv("./data/white_black_round.csv")
white_black_round_df.name = "./figures/White_Black_Round.png"

white_black_gray_round_df = pd.read_csv("./data/white_black_and_gray_round.csv")
white_black_gray_round_df.name = "./figures/White_Black_Gray_Round.png"

include_round_plots = {
    black_round_df.name: [True, False, False],
    white_round_df.name: [False, True, False],
    white_black_round_df.name: [True, True, False],
    white_black_gray_round_df.name: [True, True, True]
}

round_world_dfs = [black_round_df, white_round_df, white_black_round_df, white_black_gray_round_df]

for df in round_world_dfs:
    create_prop_temp_lum_graph(df, include_round_plots[df.name], df.name)
```
