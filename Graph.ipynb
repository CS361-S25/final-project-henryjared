```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/arroyoruizj/Desktop/ALDE_Work/final-project-henryjared/black_and_white.csv")

df = df[(df['t'] >= 50000) & (df['t'] <= 6050000)]
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Top plot: Proportion of Daisies
axs[0].plot(df['L'], df['a_b'], label='Black Daisies', color='black')
axs[0].plot(df['L'], df['a_w'], label='White Daisies', color='black', linestyle=':')

axs[0].set_xlim([0, 2])
axs[0].set_ylim([0, 0.8])
axs[0].set_yticks([.2, .4, .6])
axs[0].legend(loc='upper center', ncols=2)
axs[0].set_ylabel("Area %", size=13)


axs[1].plot(df['L'], df['temp'], color='black')
axs[1].set_xlim([0, 2])
axs[1].set_ylim([0, 70])
axs[1].set_yticks([20, 40, 60])
axs[1].set_xlabel("Solar Luminosity", size=13)
axs[1].set_ylabel("Temperature (C)", size=13)


plt.tight_layout()
plt.xlim([0.6, 1.7])
plt.savefig("BWArea_Luminosity.png")
plt.show()
```
