```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/arroyoruizj/Desktop/ALDE_Work/final-project-henryjared/black_and_white.csv")

df = df[(df['t'] >= 50000) & (df['t'] <= 6050000)]
```

```python
fig, axs = plt.subplots(2, 1, figsize=(7, 8), sharex=True)

# Top plot: Proportion of Daisies
axs[0].plot(df['L'], df['a_b'], label='Black Daisies', color='black')
axs[0].plot(df['L'], df['a_w'], label='White Daisies', color='red')

# axs[0].scatter(df['L'], df['a_b'], label='Black Daisies', c='black', marker='o')
# axs[0].scatter(df['L'], df['a_w'], label='White Daisies', c='white', edgecolors='black')

axs[0].set_xlim([0, 2])
axs[0].set_ylabel("Area %", size=13)

# Bottom plot: Temperature
# sc = axs[1].scatter(df['L'], df['temp'], c=df['temp'], cmap='coolwarm')
axs[1].plot(df['L'], df['temp'])
# axs[0].plot(df['L'], df['temp'], label='White Daisies', color='red', marker='o')
axs[1].set_xlim([0, 2])
axs[1].set_xlabel("Solar Luminosity", size=13)
axs[1].set_ylabel("Temperature (C)", size=13)

plt.tight_layout()
plt.show()
plt.savefig("BWArea_Luminosity.png")
```
