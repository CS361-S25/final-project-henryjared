```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pl1_gray_df = pd.read_csv("/Users/arroyoruizj/Desktop/ALDE_Work/final-project-henryjared/gray.csv")
pl1_gray_df = pl1_gray_df[(pl1_gray_df['t'] <= 605000)]

fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Top plot: Proportion of Daisies
axs[0].plot(pl1_gray_df['L'], pl1_gray_df['a_g'], label='Gray Daisies', color='black', linestyle='dashdot')

axs[0].set_xlim([0, 2])
axs[0].set_ylim([0, 0.8])
axs[0].set_yticks([.2, .4, .6])
axs[0].legend(loc='upper center', ncols=2)
axs[0].set_ylabel("Area %", size=13)


axs[1].plot(pl1_gray_df['L'], pl1_gray_df['temp'], color='black')
axs[1].set_xlim([0, 2])
axs[1].set_ylim([0, 70])
axs[1].set_yticks([20, 40, 60])
axs[1].set_xlabel("Solar Luminosity", size=13)
axs[1].set_ylabel("Temperature (C)", size=13)

plt.tight_layout()
plt.xlim([0.6, 1.7])
plt.show()
```
